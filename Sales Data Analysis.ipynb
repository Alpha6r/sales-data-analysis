```python
# Import Libraries & Load Data
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('sales_data.csv', parse_dates=['Date'])
df.head()
```

```python
# Basic Info and Cleanup
# Basic info
df.info()

# Check for missing values
df.isnull().sum()
```

```python
# Summary Statistics
# Descriptive statistics
df.describe()
```

```python
# Total Sales by Product
product_sales = df.groupby('Product')['Total Sales'].sum().sort_values(ascending=False)
product_sales.plot(kind='bar', title='Total Sales by Product', color='skyblue')
plt.ylabel('Sales Amount ($)')
plt.tight_layout()
plt.show()
```

```python
# Sales by Region
region_sales = df.groupby('Region')['Total Sales'].sum().sort_values(ascending=False)
region_sales.plot(kind='bar', title='Total Sales by Region', color='salmon')
plt.ylabel('Sales Amount ($)')
plt.tight_layout()
plt.show()
```

```python
# Monthly Sales Trends
# Add month column
df['Month'] = df['Date'].dt.to_period('M')

monthly_sales = df.groupby('Month')['Total Sales'].sum()
monthly_sales.plot(kind='line', marker='o', title='Monthly Sales Trend')
plt.ylabel('Sales Amount ($)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Heatmap: Product vs Region
# Pivot for heatmap
heatmap_data = df.pivot_table(index='Product', columns='Region', values='Total Sales', aggfunc='sum')
sns.heatmap(heatmap_data, annot=True, fmt='.0f', cmap='YlGnBu')
plt.title('Sales Heatmap: Product vs Region')
plt.tight_layout()
plt.show()
```
